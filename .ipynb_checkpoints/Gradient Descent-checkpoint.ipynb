### Gradient Descent ( Descenso del Gradiente)

El descenso de gradiente es un algoritmo de optimizaci칩n utilizado para encontrar los valores de los par치metros (coeficientes) de una funci칩n (f) que minimiza una funci칩n de coste (coste).

    Funcion de coste: Es una funcion que mide el acierto de nuestro algoritmo de machine learning basado en las variables de   
                      entrada y los datos reales, esta funcion puede ser n dimensional (n = numero de variables de entrada).

    Gradiente: Es una Funcion escalar que almacena las derivadas parciales de una funcion multivariable ( f(x,y,z,...) ),   
               este vector representa la direccion de ascenso mas pronunciado (el punto maximo de la funcion).

##### Learning Rate
recordemos que el gradient desecent es un algoritmo iterativo, sabiendo esto agregamos una nuevo coeficiente llamado "learning rate" o "ratio de aprendizaje" que representa la variacion del gradiente en cada iteracion, evaluando la funcion de coste por los valores definidos por el gradiente en ese punto.


![lr2](image/gradlr.png)

- Para entederlo un poco mejor esto (imagen) los graficos de abajo te muestra una funcion con un solo parametro de entrada **X** y **Y** como dominio, lo que hacemos es primero ubicarnos en un punto aleatorio de nuestra funcion global y a partir de ahi evalular la derivada de la funcion en ese punto, con el fin de analizar la pendiente tangente a ese funcion y interpretar si en ese punto la funcion esta creciendo o decreciendo,  entonces se podria interpretar 3 posibles casos: 

### Caso 1    
                                                    learning rate = 0.005
                                                    
Podemos decir que se evalua la funcion en 50 iteraciones y se ubica en un punto minimo de la funcion donde la pendiente es nula (  f'(x) = 0  ) y posiblemente ese el minimo de mi funcion de coste.

### Caso 2
                                                    learning rate = 0.05
Al ser tan grande se le es imposible al algoritmo converger en un punto minimo entrando en un bucle.

### Caso 3
                                                     learning rate = 0.000005
El entrenamiento de tu modelo seria ineficiente tardando mas interaciones para llegar al minimo local de tu funcion de coste o quedar estancados en una meseta de la funcion de coste.

## Gradient descent CODIGO

```python
#librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

#### Definir funcion

```python
def f(x,y):  #2 variables de entrada
    return np.sin(x) + np.cos(y)

#Dominio 
x = np.linspace(-4,4,100)
y = np.linspace(-4,4,100)

#al ser una funcion con el eje z necesito las cordenadas en X y Y 
# para esto vamos a utilizar meshgrid, seria bueno que le des una mirada a la documentacion de numpy
# https://interactivechaos.com/es/manual/tutorial-de-numpy/la-funcion-meshgrid
xx ,yy = np.meshgrid(x,y)
z = f(xx,yy)
```

```python
#ploting
fig = go.Figure([
    go.Surface(x=xx, y=yy, z = z)
]) 
fig.show()
```

    Para ver las curvas de nivel ( ver la funcion solo desde arriba ), vamos a utilizar matplotlib

```python
fig2, ax2 = plt.subplots()
level_map = np.linspace(np.min(z), np.max(z), num = 100)
cp = ax2.contourf(x,y,z, levels = level_map)
fig2.colorbar(cp)
```

### Definicion de derivada
Pendiente de una recta tangente a una funcion
![der](image/derivada.png)
![der](image/derivada.gif)

```python
level_map = np.linspace(np.min(z),np.max(z),100)
plt.contourf(xx,yy,z, levels = level_map)
plt.colorbar()
plt.title("Descenso del gradiente")

p = np.random.rand(2) * 8 - 4
plt.plot(p[0], p[1], 'o', c = 'k') #defiendo componentes del vector

h = 0.0001
#learning rate
lr = 0.01

# definicion de la derivada
def derivate(cp, p):
    return (f(cp[0], cp[1]) - f(p[0],p[1]))/h

def gradient(p):
    
    #comienza con 2 componentes en 0
    grad = np.zeros(2)
    
    #interar a traves de las componentes tomando en cuenta el punto definido
    for idx, val in enumerate(p):
        cp = np.copy(p)
        cp[idx] = cp[idx] + h
        
        #calcular la derivada parcial
        dp = derivate(cp,p)
        grad[idx] = dp
    
    return grad

for i in range(1000):
    p = p - lr*gradient(p)
    if(i%10 == 0):
        plt.plot(p[0], p[1], 'o', c = 'r')
plt.plot(p[0], p[1], 'o', c = 'w')
print(p)
```
